```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import seaborn as sns
sns.set_context("paper", font_scale = 2)
sns.set_style('ticks')

def add_patch(legend):
    from matplotlib.patches import Patch
    ax = legend.axes

    handles, labels = ax.get_legend_handles_labels()
    handles.append(Patch(facecolor='white', edgecolor='black'))
    labels.append('Control Sample')

    legend._legend_box = None
    legend._init_legend_box(handles, labels)
    legend._set_loc(legend._loc)
    legend.set_title(legend.get_title().get_text())

colors = ['#7f2704', '#d94801', '#fd8d3c', '#fdd0a2']
colors = colors[::-1]

df_0_control = df[(df['JClass'] == 0) & (df['Selection'] == 'Control')]
df_0_non_control = df[(df['JClass'] == 0) & (df['Selection'] != 'Control')]
df_1_control = df[(df['JClass'] == 1) & (df['Selection'] == 'Control')]
df_1_non_control = df[(df['JClass'] == 1) & (df['Selection'] != 'Control')]
df_2 = df[df['JClass'] == 2]
df_3_control = df[(df['JClass'] == 3) & (df['Selection'] == 'Control')]
df_3_non_control = df[(df['JClass'] == 3) & (df['Selection'] != 'Control')]
df_4 = df[df['JClass'] == 4]

markersize = 15
# ssfr
plt.figure(figsize=(12,8))

plt.plot(np.log10(df_0_non_control['best.stellar.m_star']), np.log10(df_0_non_control['sfr']/df_0_non_control['best.stellar.m_star']), label='JClass 0', markerfacecolor='gray',marker='o',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.6)
plt.plot(np.log10(df_0_control['best.stellar.m_star']), np.log10(df_0_control['sfr']/df_0_control['best.stellar.m_star']), markerfacecolor='gray',marker='s',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.6)
plt.plot(np.log10(df_1_non_control['best.stellar.m_star']), np.log10(df_1_non_control['sfr']/df_1_non_control['best.stellar.m_star']), label='JClass 1', markerfacecolor=colors[0], marker='o',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.8)
plt.plot(np.log10(df_1_control['best.stellar.m_star']), np.log10(df_1_control['sfr']/df_1_control['best.stellar.m_star']), markerfacecolor=colors[0], marker='s',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.8)
plt.plot(np.log10(df_2['best.stellar.m_star']), np.log10(df_2['sfr']/df_2['best.stellar.m_star']), label='JClass 2', markerfacecolor=colors[1], marker='o',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.8)
plt.plot(np.log10(df_3_non_control['best.stellar.m_star']), np.log10(df_3_non_control['sfr']/df_3_non_control['best.stellar.m_star']), label='JClass 3', markerfacecolor=colors[2], marker='o',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.8)
plt.plot(np.log10(df_3_control['best.stellar.m_star']), np.log10(df_3_control['sfr']/df_3_control['best.stellar.m_star']), markerfacecolor=colors[2], marker='s',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.8)
plt.plot(np.log10(df_4['best.stellar.m_star']), np.log10(df_4['sfr']/df_4['best.stellar.m_star']), label='JClass 4', markerfacecolor=colors[3], marker='o',markeredgecolor='k',markeredgewidth=1.3,markersize=markersize,linestyle='none',zorder=1,alpha=0.8)

plt.xlabel(r'$\log_{10} (\mathrm{M_{\star} \ (M_{\odot})})$', fontsize=30)
plt.ylabel(r'$\log_{10} (\mathrm{sSFR \ (yr^{-1})})$', fontsize=30)

plt.tick_params(axis="x", labelsize=24)
plt.tick_params(axis="y", labelsize=24)
lgnd = plt.legend(loc='lower left', prop={'size':20}, numpoints=1)
lgnd.legendHandles[0]._sizes = [markersize]
lgnd.legendHandles[1]._sizes = [markersize]
lgnd = add_patch(lgnd)
plt.grid()

# red_patch = mpatches.Patch(color='red', label='Control sample')
# plt.legend(handles=[red_patch])

plt.savefig('sSFR_vs_mass_NJ_JClass3_4_allJclasses.png', bbox_inches='tight', dpi=250)
# plt.yscale('log'); plt.xscale('log')
```
